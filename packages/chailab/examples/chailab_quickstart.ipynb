```python
import chailab as cl
from chailab.ui import Input, Slider, Label, Card, Button

# Example function

def greet(name: str, intensity: float):
    return "Hello, " + name + "!" * int(float(intensity))

demo_simple = cl.Interface(
    fn=greet,
    inputs=["text", "slider"],
    outputs=["text"],
    title="ChaiLab Quickstart (simple)",
    description="Enter your name and adjust the intensity."
)

demo_ui = cl.Interface(
    fn=greet,
    inputs=[
        Input(placeholder="Your name", label="Name", required=True),
        Slider(value=2, min=1, max=5, step=1, label="Intensity")
    ],
    outputs=["text"],
    title="ChaiLab Quickstart (ui components)",
)

# To launch, run one of the following in a Python file or notebook cell:
# demo_simple.launch()
# demo_ui.launch()
```
